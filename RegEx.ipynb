# Extracting Sentences where the word 'Love' is used

```python
import re
with open("miracle_in_the_andes.txt", "r", encoding='utf-8') as file:
    the_book = file.read()
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, the_book)
```

# Rendering the most used words in the book

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, the_book.lower())
```

```python
d= {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d_list = [(value - 1, name) for (name, value) in d.items()]
sorted(d_list, reverse=True)
```

# Load the Book

```python
import re
with open("miracle_in_the_andes.txt","r", encoding='utf-8') as file:
    book = file.read()
```

# Extract the Paragraphs where 'love' is used

```python
pattern = re.compile("[^\n]+love [^\n]+")
findings = re.findall(pattern, book)
```

```python
findings
```

# Extract Chapter Titles

```python
pattern = re.compile("[a-zA-Z ]+\n\n")
findings = re.findall(pattern, book)
```

```python
findings
```

# Function to find the Occurence of any Word

```python
def ask_me(w):
    design = re.compile('[a-zA-Z]+')
    findings = re.findall(design, book.lower())
    d = {}
    for word in findings:
        if word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
    try:
        return d[w]
    except:
        return f'The Word "{w}" does not exist! Try Again!'
```

```python
ask_me('brother')
```

```python
ask_me('love')
```

```python
ask_me('shrine')
```

```python
ask_me('hate')
```
